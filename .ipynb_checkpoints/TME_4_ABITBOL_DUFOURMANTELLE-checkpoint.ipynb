```python
def load_usps(fn):
    with open(fn,"r") as f:
        f.readline()
        data = [[float(x) for x in l.split()] for l in f if len(l.split())>2]
    tmp=np.array(data)
    return tmp[:,1:],tmp[:,0].astype(int)

def get_usps(l,datax,datay):
    if type(l)!=list:
        resx = datax[datay==l,:]
        resy = datay[datay==l]
        return resx,resy
    tmp =   list(zip(*[get_usps(i,datax,datay) for i in l]))
    tmpx,tmpy = np.vstack(tmp[0]),np.hstack(tmp[1])
    return tmpx,tmpy

def show_usps(data):
    plt.imshow(data.reshape((16,16)),interpolation="nearest",cmap="gray")
```

# Perceptron et classe Linéaire


```python
def perceptron_loss(w,x,y):
    y = y.reshape(-1,1)
    w = w.reshape(-1,1)
    return np.maximum(0, -y*np.dot(x,w))
```

```python
def perceptron_grad(w,x,y):
    y = y.reshape(-1,1)
    w = w.reshape(-1,1)
#     return np.where(-y*np.dot(x,w)< 0, 0, -y*x)
    return -(np.where(perceptron_loss(w,x,y) > 0,1,0))*y*x
```

```python
class Lineaire(object):
    def __init__(self,loss=perceptron_loss, loss_g=perceptron_grad, max_iter=1000,eps=0.01):
        self.max_iter, self.eps = max_iter,eps
        self.w = np.random.rand(datax.shape[1],1)
        self.loss,self.loss_g = loss,loss_g
        
    def fit(self,datax,datay):
        list_w = []
        list_errors = []
        for i in range(self.max_iter) :
            value = np.mean(self.loss_g(self.w,datax,datay),axis=0).reshape(-1,1)
            self.w = self.w - self.eps * value
            list_w.append(self.w)
            list_errors.append(self.loss(self.w,datax,datay).sum())

        return self.w, list_errors, list_w

    def predict(self,datax):
        return np.sign(np.dot(datax,self.w))

    def score(self,datax,datay):
        return (self.predict(datax) == datay).mean()
```

# Mini-batch et descente stochastique


```python
#maj de la classe lineaire avec mini batch et stochastique
class Lineaire(object):
    def __init__(self,loss=perceptron_loss, loss_g=perceptron_grad, max_iter=1000,eps=0.01):
        self.max_iter, self.eps = max_iter,eps
        self.w = np.random.rand(datax.shape[1],1)
        self.loss,self.loss_g = loss,loss_g
        
    def fit(self,datax,datay, choix ='batch'):
        list_w = []
        list_errors = []
        mini_batchs = 10
        for i in range(self.max_iter) :
            if choix == 'batch':
                value = np.mean(self.loss_g(self.w,datax,datay),axis=0).reshape(-1,1)
                self.w = self.w - self.eps * value
                
            if choix == 'stochastique':
                s = np.random.randint(0, len(datax)-1)
                x = datax[s]
                y = datay[s]
                self.w = self.w - self.eps * self.loss_g(self.w, x, y).reshape(-1,1)
                
            if choix == 'mini-batch':
                inds = [np.random.randint(0,len(datax)-1) for j in range(mini_batchs)]
                value = np.mean( [ self.loss_g( self.w, datax[j], datay[j]) for j in inds], axis = 0 )
                self.w = self.w - self.eps * value.reshape(-1,1)
                
            list_w.append(self.w)
            list_errors.append(self.loss(self.w,datax,datay).sum())
        return self.w, list_errors, list_w

    def predict(self,datax):
        return np.sign(np.dot(datax,self.w))

    def score(self,datax,datay):
        return (self.predict(datax) == datay).mean()
```

```python
def perceptron_choix(neg = 6, choix = 'batch',max_iterations =1000):
    uspsdatatrain = "../data/USPS_train.txt"
    uspsdatatest = "../data/USPS_test.txt"
    alltrainx,alltrainy = load_usps(uspsdatatrain)
    alltestx,alltesty = load_usps(uspsdatatest)

    datax,datay = get_usps([neg],alltrainx,alltrainy)
    testx,testy = get_usps([neg],alltestx,alltesty)

    datay = np.where(datay == neg, -1, 1)
    testy = np.where(testy == neg, -1, 1)
    
    model = Lineaire(max_iter = max_iterations)
    w, liste_error, liste_w = model.fit(datax,datay,choix)
    print("score sur le train : ", model.score(datax,datay))
    print("score sur le test : ", model.score(testx,testy))
    
    return model , w, liste_error, liste_w
```

# Projections et pénalisation


```python
def proj_poly(datax):
    proj = []
    for i in range(len(datax)):
        x = datax[i]
        tmp = [1]
        for xi in x:
            tmp += np.dot(xi, x).tolist()
        proj.append(tmp)
        
    return np.array(proj)
```

```python
def proj_biais(datax):
    return np.hstack( ( np.ones((datax.shape[0], 1)), datax ) )
```

```python
class Lineaire(object):
    def __init__(self,loss=perceptron_loss, loss_g=perceptron_grad, max_iter=1000,eps=0.01, proj = None):
        self.max_iter, self.eps = max_iter,eps
        self.proj = proj 
        self.w = np.random.rand(datax.shape[1],1)
        self.loss,self.loss_g = loss,loss_g
        
    def fit(self,datax,datay):
        if self.proj != None:
            self.datax = self.proj(datax)
        else:
            self.datax = datax
        
        list_w = []
        list_errors = []
        for i in range(self.max_iter) :
            value = np.mean(self.loss_g(self.w,datax,datay),axis=0).reshape(-1,1)
            self.w = self.w - self.eps * value
            list_w.append(self.w)
            list_errors.append(self.loss(self.w,datax,datay).sum())

        return self.w, list_errors, list_w

    def predict(self,datax):
        if self.proj != None:
            self.datax = self.proj(datax)
        else:
            self.datax = datax
            
        return np.sign(np.dot(datax,self.w))

    def score(self,datax,datay):
        return (self.predict(datax) == datay).mean()
```
